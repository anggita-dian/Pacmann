## Boolean Operations



### Comparison

* Nilai boolean dapat diperoleh melalui tes logika
* Dalam Python, tes logika ini difasilitasi oleh operator *comparison*

| Operator                                       | Sintaks|
|------------------------------------------------|--------|
| dan                                            | `and`  |
| atau                                           | `or`   |
| bukan                                          | `not`  |
| lebih kecil dari                               | <      |
| lebih besar dari                               | >      |
| lebih kecil atau sama dengan                   | <=     |
| lebih besar atau sama dengan                   | >=     |
| cek kesamaan                                   | ==     |
| cek ketidaksamaan                              | !=     |

* Dalam sesi ini, kita akan mencoba membandingkan beberapa objek dalam Python. 
* Namun akan lebih banyak digunakan objek angka untuk membahas topik ini.

Operator komparasi ini akan sangat berguna jika kita ingin mengecek suatu pernyataan apakah benar atau salah.

Misalkan, Anto dan Adi sama-sama membeli permen cokelat, dengan banyaknya sebagai berikut:

```python
permen_anto = 10
permen_adi = 9
```

permen_anto = 10
permen_adi = 9

**Pertama**, kita ingin membandingkan apakah permen yang dimiliki oleh Anto sama dengan yang dimiliki Adi. Caranya dengan menggunakan operator `==`.

```python
permen_anto == permen_adi
```

Apa maksud dari output tersebut? 

**Artinya**: pernyataan bahwa banyak permen yang dimiliki Anto sama dengan yang dimiliki Adi adalah **salah**.

Karena memang banyak permen yang dimiliki Anto dan Adi **berbeda**.

```python
permen_anto != permen_adi
```

Selanjutnya, kita ingin mengecek apakah banyak permen Anto lebih besar dari Adi?

```python
permen_anto > permen_adi
```

Tentu, karena permen Anto lebih banyak dibandingkan Adi maka pernyataan tersebut adalah benar.

Sehingga, pernyataan bahwa permen Anto lebih sedikit dari Adi adalah salah.


```python
permen_anto < permen_adi
```

### Boolean logic

Sebagai permulaan, operasi dasar boolean ada 3 yaitu:
* AND
* OR
* NOT



---
#### AND
Secara sederhana, `and` memiliki karakteristik sebagai berikut:
* Nilai `TRUE` hanya bisa diperoleh jika dua nilai bernilai `TRUE`
* Secara fisik adalah irisan antara dua kondisi.
* Contoh:
    * Ibu membeli tomat dan jeruk.
    * Maka agar kalimat tersebut bernilai benar, Ibu harus membeli tomat dan jeruk.
    * Jika Ibu hanya membeli tomat, maka kalimat tersebut akan bernilai salah.

Truth table:

| Condition 1 | Condition 2 | AND  |
|-------------|-------------|------|
| TRUE        | TRUE        | TRUE |
| TRUE        | FALSE       | FALSE|
| FALSE       | TRUE        | FALSE|
| FALSE       | FALSE       | FALSE|

#### AND

* Contoh operasi boolean pada `and`





```python
True and True
```

```python
True and False
```

```python
3 < 4 and 3 > 1
# True and True
```

```python
3 < 4 and 3 < 2
# True and False
```



---
#### OR
Secara sederhana, `or` memiliki karakteristik sebagai berikut:
* Nilai `TRUE` bisa diperoleh jika salah satu kondisi bernilai `TRUE`
* Secara fisik, `or` adalah gabungan antara dua kondisi.
* Contoh:
    * Ibu membeli tomat atau jeruk.
    * Maka, jika Ibu membeli tomat saja, kalimat tersebut sudah benar.
    * Sama halnya jika Ibu membeli jeruk saja.
    * Kalimat tersebut akan salah jika Ibu tidak membeli keduanya.

Truth table:

| Condition 1 | Condition 2 | OR   |
|-------------|-------------|------|
| TRUE        | TRUE        | TRUE |
| TRUE        | FALSE       | TRUE |
| FALSE       | TRUE        | TRUE |
| FALSE       | FALSE       | FALSE|

---
### OR

* Contoh operasi boolean pada `or`



```python
True or False
```

```python
False or False
```

```python
3 < 4 or 3 < 2
```

```python
1 > 2 or 3 < 2
```
