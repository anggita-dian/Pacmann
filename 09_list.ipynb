### Perbedaan List dengan Tuple 

* Sampai sejauh ini dapat dilihat bahwa list sangat mirip dengan tuple
* Namun, perbedaannya adalah:
    * Elemen pada list dapat diubah
    * Namun tuple tidak

```python
list_1 = [1,2,3,4,5]
list_1[1] = 55
print(list_1)
```

```python
tuple_1 = (1,2,3,4,5)
tuple_1[1] = 55
print(tuple_1)
```

## Manipulasi Elemen dalam List

**1. Menambahkan elemen pada list**












```python
list_1 = [1,2,3,4]
```

```python
print (list_1)
```

```python
list_1.append("Fifth element")
print (list_1)
```

```python
list_1.insert(0,100) # 
print(list_1)
```

```python
list_1.extend([4, 1, 6, 3, 4])
list_1
```

```python
print(list_1)
```
