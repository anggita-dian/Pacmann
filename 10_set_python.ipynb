## Set 
* Set adalah bentuk dari struktur data yang dapat memiliki beberapa elemen
* Elemen dalam set tidak disimpan berdasarkan urutan
* Set hanya memiliki elemen unik
* Untuk membuat set digunakan tanda `{}`

```python
set_1 = {'mangga','jambu','salak','kedondong'}
print(set_1)
```

```python
set_2 = {2,2,3,5,7,8,8,9,10,10,10}
print(set_2)
```

### Menambahkan elemen kedalam set



```python
thisset = {"apple", "banana", "cherry"}
```

```python
thisset.add("orange")
print(thisset)
```

### Menghapus elemen dari set 

```python
thisset = {"apple", "banana", "cherry"}
```

```python
thisset.remove("apple")
print(thisset)
```

### Mengakses elemen set 

* Elemen pada set tidak disimpan berdasarkan urutan.
* Sehingga tidak dapat dipanggil dengan menggunakan index.
* Untuk memanggil elemen set digunakan loop.



```python
set_sayur = {'sawi','brokoli','bayem','kangkung','kubis'}
```

```python
for sayur in set_sayur:
    print(sayur)
```
