## Dictionary
* Dictionary adalah struktur data yang terdiri dari pasangan _key_ dan _value_.
* Key dan value dipisahkan dengan tanda `:`
* Dictionary direpresentasikan dengan tanda `{}`



```python
orang = {'Nama':'Adi','Umur':30,'Pekerjaan':'Karyawan swasta','Domisili':'Tangerang'}
```

```python
print(orang)
```

### Mengakses Nilai dalam dictionary 

```python
customer = {'Name' : 'M. Rudi', 'Address' : 'Depok', 'Institution' : 'PACMANN AI' }
```

```python
customer['Name']
```

```python
customer['Address']
```

```python
customer['Institution']
```

## Mengakses key dan value



```python
customer.keys()
```

```python
customer.values()
```

### Menambahkan elemen



```python
customer['umur'] = 27
```

```python
print(customer)
```

```python
customer['Name'].append('Adit')
```

```python
customer = {'Name' : ['M. Rudi'], 'Address' : ['Depok'], 'Institution' : ['PACMANN AI'], 'Age':[27] }
```

```python
customer['Name']
```

```python
customer['Name'].append('Handoko')
customer['Address'].append('Malang')
customer['Institution'].append('PACMANN AI')
customer['Age'].append(32)
customer
```

```python
customer.values()
```
