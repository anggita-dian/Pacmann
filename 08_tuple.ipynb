## Tuple

* Tuple adalah struktur data yang terdiri dari beberapa elemen dan memiliki urutan.
* Tuple direpresentasikan dengan tanda kurung `()`

```python
simple_tuple = (1,3,5,7,9)
```

```python
urutan_1 = simple_tuple[0]
urutan_2 = simple_tuple[1]
urutan_3 = simple_tuple[2]
```

```python
print(urutan_1)
```

### Mengambil beberapa elemen

```python
band = ("AC/DC","Dreamtheater","Metallica","Slipknot","The Police","Led Zeppelin","Pink Floyd","Genesis")
print(band[0:4])
```

### Tuple untuk loop

```python
band = ("AC/DC","Dreamtheater","Metallica","Slipknot","The Police","Led Zeppelin","Pink Floyd","Genesis")
for nama in band:
    print(nama)
```

### Dimana Tuple dipakai? 







```python
def operation(number_1, number_2): #input or argument
    multiply = number_1 * number_2
    addition = number_1 + number_2
    return multiply, addition

result_return = operation(4, 3)
result_return
```

```python
type(result_return)
```
